## 배열 합치기
- concatenate 함수
- axis 방향으로 

```python
arr1 = np.array([
    [1,2],
    [3,4],
    [5,6]
])

arr2 = np.array([
    [10,20]
])
np.concatenate([arr1,arr2],axis=0)
```

## stack 함수
- 함치련는 차원의 크기가 모두 동일해야함
- 합치려는 배열리 하나의 행으로 취급됨

```python
arr1 = np.array([
    [1,2],
    [3,4],
    [5,6]
])

arr2 = np.array([
    [10,20]
])
np.stack([arr1,arr2],axis = 0)
```

```python
arr2 = np.array([
    [10,20],
    [30,40],
    [50,60]
])


np.stack([arr1,arr2])
```

```python
np.stack([arr1,arr2]).shape
```
